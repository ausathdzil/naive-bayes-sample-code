```python
import pandas as pd

df = pd.read_excel('../raw/prob_olahraga.xlsx')
print(df.head())
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

weather_encoded = le.fit_transform(df.Cuaca)
temp_encoded = le.fit_transform(df.Temperatur)
wind_speed_encoded = le.fit_transform(df['Kecepatan Angin'])
label = le.fit_transform(df.Berolahraga)

# print(weather_encoded)
# print(temp_encoded)
# print(wind_speed_encoded)
# print(label)

features=list(zip(weather_encoded, temp_encoded, wind_speed_encoded))
# print(features)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(features, label)

prediction = model.predict([[0, 0, 0]]) # 0: Cerah, 0: Dingin, 0: Tidak Berangin
# print(prediction)

if (prediction[0] == 0):
    print("The prediction result is: Tidak (0)")
else:
    print("The prediction result is: Ya (1)")
```
