```python
import pandas as pd

df = pd.read_csv('../raw/kelulusan_mahasiswa.csv')
df.head()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

gender_encoded = le.fit_transform(df['JENIS_KELAMIN'])
status_encoded = le.fit_transform(df['STATUS_MAHASISWA'])
marriage_encoded = le.fit_transform(df['STATUS_PERNIKAHAN'])
label = le.fit_transform(df['STATUS_KELULUSAN'])

gpa = df['IPK_Semester_1-6']

features = list(zip(gender_encoded, status_encoded, marriage_encoded, gpa))
features
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(features, label)

prediction = model.predict([[0, 1, 0, 2.7]]) # 0: LAKI-LAKI, 1: MAHASISWA, 0: BELUM, 2.7: IPK_Semester_1-6
# print(prediction)

if (prediction[0] == 0):
    print("The prediction result is: TERLAMBAT (0)")
else:
    print("The prediction result is: TEPAT (1)")
```
